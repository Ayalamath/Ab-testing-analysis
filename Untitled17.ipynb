```python
#Построить гистограммы распределения количества заказов по группам (control и test)
control_events= df[df['group']=='control']['events']
test_events = df[df['group']=='test']['events']
plt.figure(figsize = (10,5))
plt.hist(control_events, bins = 30, alpha = 0.6,edgecolor = 'black', label = 'Control')
plt.hist(test_events, bins= 30, alpha = 0.6, edgecolor = 'black',label= 'Test')
plt.xlabel('Number of events')
plt.ylabel('Frequency')
plt.title('Distribution of events by group')
plt.legend()

plt.show()
```

```python

#Построить боксплоты (box plots) для наглядного сравнения распределений в группах
control_events = df[df['group'] == 'control']['events']
test_events = df[df['group'] == 'test']['events']

plt.figure(figsize=(8,5))

plt.boxplot(
    [control_events, test_events],
    labels=['Control', 'Test'],
    showfliers=True
)

plt.ylabel('Number of events')
plt.title('Box plot of events by group')

plt.show()
```

```python
# Проверка нормальности распределения данных.
#Использовать тест Шапиро-Уилка для проверки нормальности распределения количества заказов в каждой группе.
from scipy.stats import shapiro

control_events = df[df['group'] == 'control']['events']
test_events = df[df['group'] == 'test']['events']

stat_control, p_control = shapiro(control_events)
stat_test, p_test = shapiro(test_events)

print(f'Control group: p-value = {p_control}')
print(f'Test group: p-value = {p_test}')
```

```python
### P-value меньше 0.05, значит данные не нормально распределены. Пользуемся тестом Мана Уитней
```

```python
# Построить график визуальной оценки нормальности распределения.

import scipy.stats as stats

control_events = df[df['group'] == 'control']['events']
test_events = df[df['group'] == 'test']['events']

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
stats.probplot(control_events, dist="norm", plot=plt)
plt.title('Q-Q plot: Control group')

plt.subplot(1,2,2)
stats.probplot(test_events, dist="norm", plot=plt)
plt.title('Q-Q plot: Test group')

plt.tight_layout()
plt.show()
```

```python
#Выбрать подходящий статистический тест:

from scipy.stats import mannwhitneyu

control_events = df[df['group'] == 'control']['events']
test_events = df[df['group'] == 'test']['events']

stat, p_value = mannwhitneyu(
    control_events,
    test_events,
    alternative='two-sided'
)

print(f"Mann–Whitney U statistic: {stat}")
print(f"p-value: {p_value}")
```

```python
## p -value меньше 0.05 ,нулевая гипотеза отвергается , есть значимая разница между группами.
# Проверка нормальности показала,
#что данные в контрольной и тестовой группах не имеют нормального распределения. 
#В связи с этим был применён тест Манна–Уитни. Результаты теста выявили статистически значимую разницу между группами (p-value < 0.05),
#что свидетельствует об эффективности изменений в тестовой группе.
```
